Logic 

1. Understanding the Financial goals of the user
   Objective is to determine if the user on track to meet their financial goals by the target year , given their : Current Savings for the goal , Monthly savings capacity, Goal Amount and Deadline

```python
from datetime import datetime
import random

#we need to know the current year , so to find how much years are left till goal year 
current_year= datetime.now().year

#function to evaluate goal progress
def track_goal(user):

    #1. Reading the goal info from the user data
    goal_amt = user['GoalAmount']
    deadline = user['GoalDeadlineYear']
    years_left = deadline - current_year

    #2. Calculating the user's Monthly savings and its capacity
    current_savings = user['CurrentSavingsForGoal']
    income = user['MonthlyIncome']
    expenses = user['MonthlyExpenses']
    monthly_savings = income - expenses

    #3. Now we are estimating how much will the user save in future
    future_savings = monthly_savings*12*years_left 

    #4. Estimating how current savings growth or downfall , random interest has been taken
    interest_rate = random.uniform(-0.03,0.08)  
    future_value_current = current_savings *((1+ interest_rate)**years_left)

    #5.Adding both to get the total projected amount
    total_funds = future_savings+future_value_current

    #6. Comparing the goal if it is on right track or not
    if total_funds>=goal_amt:
        suggestion = "You are on track to meet your goal"
    else:
        shortfall = goal_amt - future_value_current
        extra_needed = shortfall / (12*years_left)
        suggestion = f"You need to save about ₹{extra_needed:.0f} more every month"

   # Return the results
    return {
        "UserID": user["UserID"],
        "Goal": user["GoalName"],
        "GoalAmount": goal_amt,
        "GoalDeadline": deadline,
        "YearsLeft": years_left,
        "MonthlySavingCapacity": monthly_savings,
        "UsedInterestRate (%)": round(interest_rate * 100, 2),
        "FutureValueOfCurrentSavings": round(future_value_current, 2),
        "ProjectedFutureSavings": round(future_savings, 2),
        "TotalProjectedFunds": round(total_funds, 2),
        "Suggestion": suggestion
    }
        
```

2. Analyzing and stimulating the portfolio
   Objective : To analyze user's investment portolio and stimulate its future growth, and possibly rebalancing to meet financial goal more effectively

```python
def analyze_portfolio(user):
    stocks= user['Portfolio_Stocks']
    crypto= user['Portfolio_Crypto']
    gold= user['Portfolio_Gold']
    fd= user['Portfolio_Fixed Deposit']

    #Calculating total current value to get the total portfolio value 
    total_current = stocks+crypto+gold+fd

    #We already have current year and years left
    deadline = user['GoalDeadlineYear']
    years_left = deadline - current_year
    #So now expected return rates for each assets
    stock_rate = 0.10
    crypto_rate = random.uniform(-0.10,0.30) #Crypto is very risky
    gold_rate = 0.06
    fd_rate = 0.05

    #Next finding the future value using Compound interest
    future_stocks = stocks*((1+stock_rate)**years_left)
    future_crypto = crypto*((1+crypto_rate)**years_left)
    future_gold = gold*((1+gold_rate)**years_left)
    future_fd = fd*((1+fd_rate)**years_left)

    #Total future value of portfolio
    total_future = future_stocks+future_crypto+future_gold+future_fd

    #Calculating percentage of risky assests(stocks and crypto)
    risky_percent = ((stocks+crypto)/total_current)*100

    #Giving suggestions based on the risk
    if risky_percent>60:
        risk_level = "High"
        suggestion = "Too much in stocks and crypto. Consider moving to some other assest"
    elif risky_percent >30:
        risk_level = "Medium"
        suggestion = "Moderate risk, Stay careful if goal is near."
    else:
        risk_level = "Low"
        suggestion = "Your Portfolio is stable. Good for short-term goals."

    return {
       "UserID": user["UserID"],
        "CurrentPortfolioValue": round(total_current, 2),
        "FuturePortfolioValue": round(total_future, 2),
        "YearsToGoal": years_left,
        "RiskLevel": risk_level,
        "UsedCryptoReturnRate (%)": round(crypto_rate * 100, 2),
        "Suggestion": suggestion
   }
    
```

2. Recommends Action
   Objective: This is the AI part where personalized recommendations will be given by the system

```python
def recommend_action(goal_result, portfolio_result):
    # Step 1: Extract needed values
    gap = goal_result['GoalAmount'] - goal_result['TotalProjectedFunds']
    years_left = goal_result['YearsLeft']
    saving_capacity = goal_result['MonthlySavingCapacity']
    on_track = gap <= 0

    risk = portfolio_result['RiskLevel']
    future_value = portfolio_result['FuturePortfolioValue']
    suggestion = ""

    # Step 2: Decision logic

    #Case 1: On track and risk is low or medium
    if on_track and risk in ['Low', 'Medium']:
        suggestion = "You're on track and your portfolio is well-balanced. No changes needed."

    #Case 2: On track but portfolio is high risk
    elif on_track and risk == 'High':
        suggestion = "You're on track, but your portfolio is risky. Shift 10-20% from crypto/stocks to FD or gold."

    #Case 3: Not on track but user has good saving capacity
    elif not on_track and saving_capacity > 0:
        extra_monthly = gap / (12 * years_left)
        suggestion = (
            f"You're falling short of your goal. Increase monthly savings by ₹{extra_monthly:.0f} "
            f"and avoid risky assets like crypto."
        )

    #Case 4: Not on track and risk is high
    elif not on_track and risk == 'High':
        suggestion = (
            "You are far from your goal and your portfolio is risky. Shift some funds to safer assets "
            "like FD or gold and try to save more if possible."
        )

    # ⏳ Case 5: Time is short
    elif years_left <= 3:
        suggestion = "Goal is near. Reduce risky assets and move most of your funds to FD or gold."

    else:
        suggestion = "Review your savings and investments. Try to increase your savings or adjust your goal."

    # Step 3: Return final recommendation
    return {
        "UserID": goal_result["UserID"],
        "OnTrack": on_track,
        "GoalGap (₹)": round(gap, 2),
        "PortfolioRisk": risk,
        "YearsLeft": years_left,
        "FinalSuggestion": suggestion
    }
```
